```python
import cv2
import os

# === Configuration ===
video_path = r"F:\Output\runs\SUMIYOSHI_identification\parallel_tracking_identification\May 1st Report 2025\Sumiyoshi-44\Base_rtx8000_June_15000_v2\_Ch006_DX-FHD-5-54\2025-04-10_1505_part 20\1505_identification.mp4"
#video_path =  r"D:\Output\runs\SUMIYOSHI_identification\parallel_tracking_identification\May 1st Report 2025\Sumiyoshi-44\April_Day_Night_10000_iters_v1\_Ch006_DX-FHD-5-54\2025-04-16_1105_part 1\1105_identification.mp4"
output_folder = r'C:\NyiZaw\Report\2025\July\after-new-1505-images'
target_fps = 5

# === Create output directory ===
os.makedirs(output_folder, exist_ok=True)

# === Open the video file ===
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    raise IOError(f"Cannot open video file: {video_path}")

# Get original video FPS
original_fps = cap.get(cv2.CAP_PROP_FPS)
frame_interval = int(original_fps // target_fps)

print(f"Original FPS: {original_fps}, Extracting every {frame_interval} frames.")

frame_count = 0
saved_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Save every nth frame
    if frame_count % frame_interval == 0:
        frame_filename = os.path.join(output_folder, f"frame_{saved_count:05d}.jpg")
        cv2.imwrite(frame_filename, frame)
        saved_count += 1

    frame_count += 1

cap.release()
print(f"Done! Saved {saved_count} frames to '{output_folder}'.")
```
