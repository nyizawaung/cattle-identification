```python
import os
import shutil
import random
from pathlib import Path

def split_per_cattle(
    src_root,
    train_root,
    val_root,
    train_ratio=0.8,
    seed=42,
    min_train=3,
    min_val=2
):
    random.seed(seed)

    src_root = Path(src_root)
    train_root = Path(train_root)
    val_root = Path(val_root)

    train_root.mkdir(parents=True, exist_ok=True)
    val_root.mkdir(parents=True, exist_ok=True)

    cattle_ids = [d for d in src_root.iterdir() if d.is_dir()]

    for cow_dir in cattle_ids:
        images = list(cow_dir.glob("*"))
        n = len(images)

        if n < min_train:
            print(f"[SKIP] {cow_dir.name}: only {n} images")
            continue

        random.shuffle(images)

        n_val = max(int(n * (1 - train_ratio)), min_val)
        n_val = min(n_val, n - min_train)

        val_imgs = images[:n_val]
        train_imgs = images[n_val:]

        cow_train_dir = train_root / cow_dir.name
        cow_val_dir = val_root / cow_dir.name

        cow_train_dir.mkdir(exist_ok=True)
        cow_val_dir.mkdir(exist_ok=True)

        for img in train_imgs:
            shutil.copy2(img, cow_train_dir / img.name)

        for img in val_imgs:
            shutil.copy2(img, cow_val_dir / img.name)

        print(
            f"[OK] {cow_dir.name}: "
            f"train={len(train_imgs)}, val={len(val_imgs)}"
        )

    print("✅ Identity-aware split complete")
split_per_cattle(
    src_root=r"C:\Identification dataset\KNP_identification_dataset\RAW",
    train_root=r"C:\Identification dataset\KNP_identification_dataset\Training",
    val_root=r"C:\Identification dataset\KNP_identification_dataset\Validation",
    train_ratio=0.8
)
```
