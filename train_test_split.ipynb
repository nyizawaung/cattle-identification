```python
import json
import numpy as np
from os import rename
import cv2
import os
import glob
import shutil
import random
from detectron2.structures import BoxMode
class NumpyEncoder(json.JSONEncoder):
    """ Special json encoder for numpy types """
    def default(self, obj):
        if isinstance(obj, (np.int_, np.intc, np.intp, np.int8,
                            np.int16, np.int32, np.int64, np.uint8,
                            np.uint16, np.uint32, np.uint64)):
            return int(obj)
        elif isinstance(obj, (np.float_, np.float16, np.float32,
                              np.float64)):
            return float(obj)
        elif isinstance(obj, (np.ndarray,)):
            return obj.tolist()
        return json.JSONEncoder.default(self, obj)

    
image_path="F:\\Mounting_Model_Data\\Yoshii\\version37\\Tanaka Mounting Data"
training_images_path = f"{image_path}\\clean_images"

PATH = os.path.join(image_path,"*.json")
annotations= dict()
training_annotations = dict()

os.makedirs(training_images_path, exist_ok=True)

for jsonPath in glob.glob(PATH):
    with open(jsonPath) as f1:   #D:/Projects/WH/Python/Detectron2/Yoshii_22070506/Yoshii20220705_28_000/001_220705060000_6001_20_00_28_00/via_project_8Aug2022_11h36m_json.json
        data_json = json.load(f1)
    for a,val in enumerate(data_json.values()):
        filename = val['filename']
        filePath= os.path.join(image_path,filename)
        height,width = cv2.imread(filePath).shape[:2]
        center_x = width/2 + 1
        center_y = height/2 + 1
        width_roi = 0.5
        height_roi = 0.7
        x1 = int(center_x - (width_roi*width)/2)
        x2 = int(center_x + (width_roi*width)/2)
        y1 = int(center_x - (height_roi*height)/2)
        y2 = int(center_x + (height_roi*height)/2)
        Regions = val['regions']
        for r in range(len(Regions)):
            classId = val['regions'][r]['region_attributes']['cow']
            print(classId)
            if classId == "2":
                print(filePath)
                x_all_points = val['regions'][r]['shape_attributes']['all_points_x']
                y_all_points = val['regions'][r]['shape_attributes']['all_points_y']
                for point in range(len(x_all_points)):
                    x = x_all_points[point]
                    y = y_all_points[point]
                    if x1< x <x2 and y1 < y < y2 :
                        size = val['size']
                        annotations[f"{filename}{size}"] = val
                        break

annotationList = list(annotations.values())

 
for val in annotationList:
    filename = val['filename']
    size = val['size']
    training_annotations[f"{filename}{size}"] = val  
    shutil.copyfile(f'{image_path}/{filename}', f'{training_images_path}/{filename}')  
        
with open(f'{training_images_path}/new_dataset_clean_annotations_json.json', 'w') as outfile:
   json.dump(training_annotations, outfile, cls=NumpyEncoder) 
```
