```python
import numpy as np
from collections import defaultdict
class responseModel():
    def __init__(self,cam_counter,tracking_ids,predicted_ids):
        self.cam_counter = cam_counter
        self.tracking_ids = tracking_ids
        self.predicted_ids = predicted_ids

def abc(cam_counter,start,end):
    tracking_ids = []
    predicted_ids = []
    for i in  range(start,end):
        tracking_ids.append(12 if i==2 or i==7  else 12+i)
        predicted_ids.append(10 if i==2 or i== 7 or i==3 else 5)
        

    return responseModel(cam_counter,tracking_ids,predicted_ids)

def get_duplicate_indexes(array):
    index_map = defaultdict(list)
    for i, val in enumerate(array):
        index_map[val].append(i)

    # Only keep values with duplicates
    duplicates = {val: idxs for val, idxs in index_map.items() if len(idxs) > 1}
    return duplicates

all_result = []
all_result.append(abc(0,0,5))
all_result.append(abc(1,6,10))
all_predicted_ids = get_duplicate_indexes([item for sublist in all_result for item in sublist.predicted_ids])
all_tracking_ids = [item for sublist in all_result for item in sublist.tracking_ids]

reset_tracking_ids = []
for duplicate in all_predicted_ids:
    print(duplicate, " duplicate value" )
    duplicate_tracking_ids = [all_tracking_ids[i] for i in all_predicted_ids[duplicate]]
    print(duplicate_tracking_ids, " duplicate tracking ids")
    is_same = all(val == duplicate_tracking_ids[0] for val in duplicate_tracking_ids)
    if not is_same:
        for id in duplicate_tracking_ids:
            if id not in reset_tracking_ids:
                reset_tracking_ids.append(id)
print(reset_tracking_ids)
#print(all_predicted_ids)
#for result in all_result:
#    print(result.predicted_ids)
```

```python
import json
import csv
import os
from collections import defaultdict
from PIL import Image

# Paths
coco_json_path = r"D:\Nyi Zaw Aung\815_CowDataChecking\Sumiyoshi ToAnnotate\April 2025\NightTime\Left Lane\left_lane_cvat_coco.json"
images_dir = r"D:\Nyi Zaw Aung\815_CowDataChecking\Sumiyoshi ToAnnotate\April 2025\NightTime\Left Lane"  # Path to images if you want file size
output_csv_path = r'D:\Nyi Zaw Aung\815_CowDataChecking\Sumiyoshi ToAnnotate\April 2025\NightTime\Left Lane\left_lane_vgg_csv.csv'

# Load COCO annotations
with open(coco_json_path, 'r') as f:
    coco = json.load(f)

# Build index
image_id_to_filename = {img['id']: img['file_name'] for img in coco['images']}
image_id_to_size = {img['id']: img['width'] * img['height'] for img in coco['images']}
image_id_to_annotations = defaultdict(list)
for ann in coco['annotations']:
    image_id_to_annotations[ann['image_id']].append(ann)

# Category mapping
cat_id_to_name = {cat['id']: cat['name'] for cat in coco['categories']}

# Write CSV
with open(output_csv_path, mode='w', newline='') as csv_file:
    fieldnames = ['filename', 'file_size', 'file_attributes', 'region_count', 'region_id', 'region_shape_attributes', 'region_attributes']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()

    for image_id, anns in image_id_to_annotations.items():
        filename = image_id_to_filename[image_id]
        filepath = os.path.join(images_dir, filename)
        try:
            file_size = os.path.getsize(filepath)
        except:
            file_size = image_id_to_size[image_id]  # fallback
        region_count = len(anns)

        for region_id, ann in enumerate(anns):
            if 'segmentation' in ann and isinstance(ann['segmentation'], list):
                for seg in ann['segmentation']:
                    xs = seg[0::2]
                    ys = seg[1::2]
                    region_shape = {
                        "name": "polygon",
                        "all_points_x": xs,
                        "all_points_y": ys
                    }
                    region_attr = {
                        "label": cat_id_to_name[ann['category_id']]
                    }
                    writer.writerow({
                        "filename": filename,
                        "file_size": file_size,
                        "file_attributes": "{}",
                        "region_count": region_count,
                        "region_id": region_id,
                        "region_shape_attributes": json.dumps(region_shape),
                        "region_attributes": json.dumps(region_attr)
                    })
```

```python
import os
import cv2
import numpy as np
def unsharp_mask(bgr, radius=2.5, amount=1.0):
    blurred = cv2.GaussianBlur(bgr, (0,0), radius)
    sharp = cv2.addWeighted(bgr, 1+amount, blurred, -amount, 0)
    return np.clip(sharp, 0, 255).astype(np.uint8)

def apply_clahe_ycrcb(bgr, clip=1.5, tile=8):
    ycrcb = cv2.cvtColor(bgr, cv2.COLOR_BGR2YCrCb)
    y, cr, cb = cv2.split(ycrcb)
    clahe = cv2.createCLAHE(clipLimit=clip, tileGridSize=(tile, tile))
    y2 = clahe.apply(y)
    merged = cv2.merge([y2, cr, cb])
    return cv2.cvtColor(merged, cv2.COLOR_YCrCb2BGR)

path = "C:/Users/choni/Pictures/KNP cameras"

for file in os.listdir(path):
    if file.endswith(".png"):
        image_path = os.path.join(path, file)
        save_path = image_path.replace(".png","_clahe.png")
        orig_image = cv2.imread(image_path)
        img = orig_image.copy()
        img2 = img.copy()

        img = apply_clahe_ycrcb(img,clip=2.5, tile=8)
        img = unsharp_mask(img, radius=1.4, amount=1.0)
        
        #cv2.imwrite(image_path, img)
        img2 = apply_clahe_ycrcb(img2,clip=0.8, tile=16)
        img2 = unsharp_mask(img2, radius=1.0, amount=0.4)
   #frame = self.apply_clahe_ycrcb(frame, clip=2.5, tile=8)  #B
                #frame = self.unsharp_mask(frame, radius=1.4, amount=1.0)
        # cv2.imshow("Processed Image", img)
        # cv2.imshow("Image 2 Image", img2)
        cv2.imwrite(save_path, img)
        stacked_image = np.hstack(( img,orig_image, img2))
        h,w = stacked_image.shape[:2]
        resize = (int(w*0.5), int(h*0.5))
        stacked_image = cv2.resize(stacked_image, resize, interpolation=cv2.INTER_AREA)
        #cv2.imshow("stacked 2 Image", stacked_image)
        #cv2.waitKey(0)
cv2.destroyAllWindows()
```
